### 1. What are the most popular neighbourhoods (with respect to booked nights)?

```python
neighbourhood_grouping = df.groupby('neighbourhood')['minimum_nights'].sum().reset_index()
neighbourhood_grouping_sorted=neighbourhood_grouping.sort_values(by='minimum_nights',ascending=False)
neighbourhood_grouping_sorted
```

### From the above table we can see the (most popular neighbourhoods with respect to booked nights)

```python
plt.figure(figsize=(10,6))
bars=plt.bar(neighbourhood_grouping_sorted['neighbourhood'][:10],neighbourhood_grouping_sorted['minimum_nights'][:10],color='blue')
plt.xlabel('Neighborhoods')
plt.ylabel('Minimum nights')
plt.title('Top Neighborhoods by booked nights')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

plt.show()
```

### The above graph shows the visual representation of the most popular neighbourhoods with respect to booked nights

```python
neighbourhood_grouping_unoccupied_days = df.groupby('neighbourhood')['Unoccupied days\n(in last 365 days)'].sum().reset_index()
neighbourhood_grouping_unoccupied_days=neighbourhood_grouping_unoccupied_days.sort_values(by='Unoccupied days\n(in last 365 days)')
neighbourhood_grouping_unoccupied_days
```

```python
plt.figure(figsize=(10,6))
plt.bar(neighbourhood_grouping_unoccupied_days['neighbourhood'][:10],neighbourhood_grouping_unoccupied_days['Unoccupied days\n(in last 365 days)'][:10],color='blue')
plt.xlabel('Neighborhoods')
plt.ylabel('Unoccupied days\n(in last 365 days)')
plt.title('Top Neighborhoods by Unoccupied days\n(in last 365 days)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 2. What is the highest revenue with respect to neighbourhoods?

```python
df['revenue']=df['price per night']*df['minimum_nights']
revenue_by_neighbourhood=df.groupby('neighbourhood')['revenue'].sum().reset_index()
revenue_by_neighbourhood_sorted=revenue_by_neighbourhood.sort_values(by='revenue',ascending=False)
revenue_by_neighbourhood_sorted
```

### The above table represents the highest revenue with respect to neighbourhoods

```python
plt.figure(figsize=(10,6))
bars = plt.bar(revenue_by_neighbourhood_sorted['neighbourhood'][:10],revenue_by_neighbourhood_sorted['revenue'][:10],color='orange')
plt.xlabel('Neighbourhood')
plt.ylabel('Total Revenue by neighbourhood')
plt.title('Top Neighborhoods by Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')


plt.show()
```

### The above graphical representation shows the highest revenue with respect to neighbourhoods

```python
booking_value=0.6
reviews_value=0.4

df['score']=(df['revenue']*booking_value)+(df['number_of_reviews']*reviews_value)
best_listing=df.groupby(['id','name','neighbourhood','host_name'])['score'].sum().reset_index()
best_listing_sorted=best_listing.sort_values(by='score',ascending=False)
best_listing_sorted.head(10)
```

### The above graph shows the best listings(Top 10) with the composite score for each neighbourhood with the (Weightage - 0.6 value of booking, 0.4 number of reviews). 

### 4. To Perform Correlation Analysis : Finding out which factors is causing more booked nights


```python
df.columns
```

```python
numerical_variables = ['price per night', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'Host Listings Count','Unoccupied days\n(in last 365 days)']

correlation_matrix = df[numerical_variables].corr()
correlation_with_booked_nights = correlation_matrix['minimum_nights']
correlation_with_booked_nights_sorted = correlation_with_booked_nights.abs().sort_values(ascending=False)

print("Correlation coefficients with booked nights:")
print(correlation_with_booked_nights_sorted)
```

### From the above we can see the correlation coefficients with respect to booked nights we can see there is not much correlation with the features, the highest is for Unoccupied days\n(in last 365 days) which is 0.072641 and even that is not that high which is very minimal
 

```python
plt.figure(figsize=(9, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### The above heatmap shows the correaltion between all the numerical variables we can see from the heatmap as well there is not much correlation between different variables the highest positive correlation is for number_of_reviews and reviews_per_month which is 0.70 which makes sense wrt reviews but it does not give us any inference wrt booked nights 

### 5. Specific views on Data

### 5.1  Booking performance by room type

```python

booking_by_room_type=df.groupby('room_type')['minimum_nights'].sum().reset_index()
booking_by_room_type
```

### From the above table we can see the booking performance with respect to room type

```python
plt.figure(figsize=(6, 4))
bars=plt.bar(booking_by_room_type['room_type'], booking_by_room_type['minimum_nights'], color='skyblue')
plt.xlabel('Room Type')
plt.ylabel('Total Booked Nights')
plt.title('Booking Performance by Room Type')
plt.xticks(rotation=45)
plt.tight_layout()

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

plt.show()
```

### From the graph we can see the room type performance with respect to bookings

### 5.2   Booking performance by reviews (bucketing)


```python
df['review_bucket'] = pd.cut(df['number_of_reviews'], bins=[0, 50, 100, float('inf')], labels=['Low', 'Medium', 'High'])
booking_by_reviews = df.groupby('review_bucket')['minimum_nights'].sum().reset_index()

plt.figure(figsize=(6,4))
plt.bar(booking_by_reviews['review_bucket'], booking_by_reviews['minimum_nights'], color='purple')
plt.xlabel('Review Bucket')
plt.ylabel('Total Booked Nights')
plt.title('Booking Performance by Reviews')
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```

### From the graph we can see the booking performance by reviews with bucketing

### 5. Business done by hosts sorted in descending order i.e., by revenue

```python
business_done_by_hosts=df.groupby('host_name')['revenue'].sum().reset_index()
business_done_by_hosts_sorted=business_done_by_hosts.sort_values(by='revenue', ascending=False)
business_done_by_hosts_sorted.head(15)
```

### From the table we can see the top 10 hosts who have the business in terms of revenue in descending order

```python
plt.figure(figsize=(8,4))
bars=plt.bar(business_done_by_hosts_sorted['host_name'][:10],business_done_by_hosts_sorted['revenue'][:10],color='cyan')
plt.xlabel('host_name')
plt.ylabel('revenue')
plt.title('Business done by hosts')
plt.xticks(rotation=45)
plt.tight_layout()

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

plt.show()
```

### The graph shows the visual representation of business done by hosts in decreasing order in terms of revenue
